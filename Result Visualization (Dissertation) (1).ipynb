```python
# Function to plot data for each model

def plot_model_data(ratios, model_data, model_name, colors):
    fig, axes = plt.subplots(3, 1, figsize=(10, 18))
    
    # Plot bias levels
    axes[0].plot(ratios, model_data['bias_male'][0], marker='o', linestyle='-', color=colors[0], label='Male Priority - 50% Data')
    axes[0].plot(ratios, model_data['bias_male'][1], marker='o', linestyle='-', color=colors[1], label='Male Priority - 30% Data')
    axes[0].plot(ratios, model_data['bias_male'][2], marker='o', linestyle='-', color=colors[2], label='Male Priority - 10% Data')
    axes[0].plot(ratios, model_data['bias_female'][0], marker='x', linestyle='--', color=colors[0], label='Female Priority - 50% Data')
    axes[0].plot(ratios, model_data['bias_female'][1], marker='x', linestyle='--', color=colors[1], label='Female Priority - 30% Data')
    axes[0].plot(ratios, model_data['bias_female'][2], marker='x', linestyle='--', color=colors[2], label='Female Priority - 10% Data')
    axes[0].set_title(f'{model_name} - Bias Levels Across Different Data Ratios')
    axes[0].set_xlabel('Male to Female Ratio')
    axes[0].set_ylabel('Bias Level')
    axes[0].legend()
    axes[0].grid(True)
    
    # Plot accuracy
    axes[1].plot(ratios, model_data['accuracy_male'][0], marker='o', linestyle='-', color=colors[0], label='Male Priority - 50% Data')
    axes[1].plot(ratios, model_data['accuracy_male'][1], marker='o', linestyle='-', color=colors[1], label='Male Priority - 30% Data')
    axes[1].plot(ratios, model_data['accuracy_male'][2], marker='o', linestyle='-', color=colors[2], label='Male Priority - 10% Data')
    axes[1].plot(ratios, model_data['accuracy_female'][0], marker='x', linestyle='--', color=colors[0], label='Female Priority - 50% Data')
    axes[1].plot(ratios, model_data['accuracy_female'][1], marker='x', linestyle='--', color=colors[1], label='Female Priority - 30% Data')
    axes[1].plot(ratios, model_data['accuracy_female'][2], marker='x', linestyle='--', color=colors[2], label='Female Priority - 10% Data')
    axes[1].set_title(f'{model_name} - Accuracy Across Different Data Ratios')
    axes[1].set_xlabel('Male to Female Ratio')
    axes[1].set_ylabel('Accuracy (%)')
    axes[1].legend()
    axes[1].grid(True)
    
    # Plot precision
    axes[2].plot(ratios, model_data['precision_male'][0], marker='o', linestyle='-', color=colors[0], label='Male Priority - 50% Data')
    axes[2].plot(ratios, model_data['precision_male'][1], marker='o', linestyle='-', color=colors[1], label='Male Priority - 30% Data')
    axes[2].plot(ratios, model_data['precision_male'][2], marker='o', linestyle='-', color=colors[2], label='Male Priority - 10% Data')
    axes[2].plot(ratios, model_data['precision_female'][0], marker='x', linestyle='--', color=colors[0], label='Female Priority - 50% Data')
    axes[2].plot(ratios, model_data['precision_female'][1], marker='x', linestyle='--', color=colors[1], label='Female Priority - 30% Data')
    axes[2].plot(ratios, model_data['precision_female'][2], marker='x', linestyle='--', color=colors[2], label='Female Priority - 10% Data')
    axes[2].set_title(f'{model_name} - Precision Across Different Data Ratios')
    axes[2].set_xlabel('Male to Female Ratio')
    axes[2].set_ylabel('Precision (%)')
    axes[2].legend()
    axes[2].grid(True)
    
    plt.tight_layout()
    plt.savefig(f'{model_name}_plots.png')
    plt.show()

# Defining the colors

colors = ['blue', 'green', 'red']

# Ratios for plotting

ratios = ['90:10', '80:20', '70:30', '60:40', '50:50']
```
